```python
import pandas as pd, numpy as np

%load_ext autoreload
%autoreload 2

# Считаем крабов для примера работы рисовалки
tr = pd.read_csv('crabs_train.csv')

tr.columns = tr.columns.str.lower()
tr.rename(columns={
    'shucked weight': 'shucked_weight',
    'viscera weight': 'viscera_weight',
    'shell weight': 'shell_weight',
}, inplace=True)
tr.head(3)
```

```python
# ПРИМЕРОЧНАЯ

from utils.utils import plot_dim_reduction
from utils.reductors import *
from collections import namedtuple

hue_info = namedtuple('hue_info', ['field_name', 'is_categorical'])

mapper_dict = {
    'tsne 2D': {
        'params': {
           'n_jobs': 8,
           'verbose': False,
           'n_components': 2
        },
        'features': tr.drop(columns=['sex', 'age']).columns,
        'hue': hue_info(field_name='age', is_categorical=False),
        'func': make_tsne,
    },
    'UMAP 2D': {
        'params': {
           'n_neighbors': 11,
           'min_dist': 0.1,
           'n_jobs': 8,
           'verbose': False,
           'n_components': 2
        },
        'features': tr.drop(columns=['sex', 'age']).columns,
        'hue': hue_info(field_name='sex', is_categorical=True),
        'func': make_umap,
    },
    'PCA 2D': {
        'params': {
           'n_components': 2
        },
        'features': tr.drop(columns=['sex', 'age']).columns,
        'hue': hue_info(field_name='age', is_categorical=False),
        'func': make_pca,
    },
}

plot_dim_reduction(
    data = tr.sample(1_000),
    mapper_dict=mapper_dict,
    row_height=450
)
```
